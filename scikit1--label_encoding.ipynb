```python
# label encoding - nominal data
import pandas as pd
from sklearn.preprocessing import LabelEncoder
#sample data
data=pd.DataFrame({
    'fruit':['apple','banana','mango','apple','banana']
})
#label encoding
le=LabelEncoder()
data['fruit_en']=le.fit_transform(data['fruit'])
data
```

```python
data=pd.DataFrame({
    'city':['delhi','mumbai','chennai','delhi','chennai'],
    'gender':['male','female','female','male','female']
})
le_city=LabelEncoder()
le_gender=LabelEncoder()
data['city_en']=le_city.fit_transform(data['city'])
data['gender_en']=le_gender.fit_transform(data['gender'])
data
```

```python
#ordinal data
data=pd.DataFrame({
    'size':['small','medium','large','medium','small']
            })
le=LabelEncoder()
data['size_en']=le.fit_transform(data['size'])
data
```

```python
#wrong order
#so we use OrdinalEncoder
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
data=pd.DataFrame({
    'size':['small','medium','large','medium','small']
})
#define order
encoder=OrdinalEncoder(categories=[['small','medium','large']])
data['size_en']=encoder.fit_transform(data[['size']])
data
```

```python
#one hot encoding - categorical variables-->binary variables
import pandas as pd
data=pd.DataFrame({
    'color':['red','blue','green','red']
})
ohe=pd.get_dummies(data,columns=['color'])
ohe
```

```python
#by importing OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd
data=np.array([['red'],['blue'],['green'],['red']])
#initialize encoder
encoder=OneHotEncoder(sparse_output=False)
#fit and transform
ohe=encoder.fit_transform(data)
#get feature names for columns
columns=encoder.get_feature_names_out(['color'])
#convert to dataframe
df_ohe=pd.DataFrame(ohe,columns=columns)
df_ohe
```
